```python
from agents import Agent, Runner
from agents import enable_verbose_stdout_logging
from agents.tool import function_tool
from agents.stream_events import RawResponsesStreamEvent
from dotenv import load_dotenv
import asyncio

load_dotenv()
enable_verbose_stdout_logging()

@function_tool
def weather(location: str) -> str:
    """
    Get the current weather for a given location.
    """
    return f"The weather in {location} is sunny with a high of 75F."

async def main():

    agent = Agent(name="Assistant", instructions="You are a helpful assistant", tools=[weather])

    result = Runner.run_streamed(agent, "Write a 30line story about a cat who learns to code.")
    print("-------BELOW IS THE FINAL OUTPUT-------")
    async for event in result.stream_events():
        print(f"\nEVENT: {event}")


if __name__ == "__main__":
    asyncio.run(main())
```

```python
from agents import Agent, Runner
from agents import enable_verbose_stdout_logging
from agents.tool import function_tool
from agents.stream_events import RawResponsesStreamEvent
from dotenv import load_dotenv
import asyncio

load_dotenv()
enable_verbose_stdout_logging()

@function_tool
def weather(location: str) -> str:
    """
    Get the current weather for a given location.
    """
    return f"The weather in {location} is sunny with a high of 75F."

async def main():

    agent = Agent(name="Assistant", instructions="You are a helpful assistant", tools=[weather])

    result = Runner.run_streamed(agent, "What is the weather in Islamabad?")
    print("-------BELOW IS THE FINAL OUTPUT-------")
    async for event in result.stream_events():
        # print(f"\nEVENT: {event}")
        if event.type == "raw_response_event" and isinstance(event.data, RawResponsesStreamEvent):
            print(f"\nData: {event.data}")


if __name__ == "__main__":
    asyncio.run(main())
```
